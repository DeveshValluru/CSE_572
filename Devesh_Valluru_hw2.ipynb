```python
print("Before", train_df.shape, test_df.shape, combine[0].shape, combine[1].shape)

train_df = train_df.drop(['Ticket'], axis=1)
test_df = test_df.drop(['Ticket'], axis=1)
combine = [train_df, test_df]

"After", train_df.shape, test_df.shape, combine[0].shape, combine[1].shape
```

```python
import re

# Mapping Deck letters to numeric values
deck_mapping = {"A": 1, "B": 2, "C": 3, "D": 4, "E": 5, "F": 6, "G": 7, "U": 8}

for dataset in combine:
    # Fill missing Cabin with "U0" and ensure string type
    dataset['Cabin'] = dataset['Cabin'].fillna('U0').astype(str)

    # Extract Deck safely (take first letter, or 'U' if none)
    dataset['Deck'] = dataset['Cabin'].map(
        lambda x: re.search(r'([A-Za-z])', x).group() if re.search(r'([A-Za-z])', x) else 'U'
    )

    # Map Deck letters to numbers
    dataset['Deck'] = dataset['Deck'].map(deck_mapping).fillna(0).astype(int)
```

```python
guess_ages = np.zeros((2,3))
guess_ages
```

```python
for dataset in combine:
    for i in range(0, 2):
        for j in range(0, 3):
            guess_df = dataset[(dataset['Sex'] == i) & \
                                  (dataset['Pclass'] == j+1)]['Age'].dropna()

            # age_mean = guess_df.mean()
            # age_std = guess_df.std()
            # age_guess = rnd.uniform(age_mean - age_std, age_mean + age_std)

            age_guess = guess_df.median()

            # Convert random age float to nearest .5 age
            guess_ages[i,j] = int( age_guess/0.5 + 0.5 ) * 0.5

    for i in range(0, 2):
        for j in range(0, 3):
            dataset.loc[ (dataset.Age.isnull()) & (dataset.Sex == i) & (dataset.Pclass == j+1),\
                    'Age'] = guess_ages[i,j]

    dataset['Age'] = dataset['Age'].astype(int)

train_df.head()
```

```python
train_df['AgeBand'] = pd.cut(train_df['Age'], 5)
train_df[['AgeBand', 'Survived']].groupby(['AgeBand'], as_index=False).mean().sort_values(by='AgeBand', ascending=True)
```

```python
for dataset in combine:
    dataset.loc[ dataset['Age'] <= 16, 'Age'] = 0
    dataset.loc[(dataset['Age'] > 16) & (dataset['Age'] <= 32), 'Age'] = 1
    dataset.loc[(dataset['Age'] > 32) & (dataset['Age'] <= 48), 'Age'] = 2
    dataset.loc[(dataset['Age'] > 48) & (dataset['Age'] <= 64), 'Age'] = 3
    dataset.loc[ dataset['Age'] > 64, 'Age']
train_df.head()
```

```python
for dataset in combine:
    dataset.loc[ dataset['Fare'] <= 7.91, 'Fare'] = 0
    dataset.loc[(dataset['Fare'] > 7.91) & (dataset['Fare'] <= 14.454), 'Fare'] = 1
    dataset.loc[(dataset['Fare'] > 14.454) & (dataset['Fare'] <= 31), 'Fare']   = 2
    dataset.loc[ dataset['Fare'] > 31, 'Fare'] = 3
    dataset['Fare'] = dataset['Fare'].astype(int)

train_df = train_df.drop(['FareBand'], axis=1)
combine = [train_df, test_df]

train_df.head(10)
```

```python
X_train = train_df.drop("Survived", axis=1)
Y_train = train_df["Survived"]
X_test  = test_df.drop("PassengerId", axis=1).copy()
X_train.shape, Y_train.shape, X_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
from sklearn.model_selection import cross_val_score, GridSearchCV
```

```python
param_grid = {
    'max_depth' : [2,3,5,10,20],
    'min_samples_leaf' : [1, 2, 4],
    'min_samples_split' : [2,5,10,20]

}

decison_tree = DecisionTreeClassifier()
grid_search = GridSearchCV(estimator=decison_tree, param_grid=param_grid, cv=5, scoring= 'accuracy')
grid_search.fit(X_train, Y_train)

best_desicion_tree = grid_search.best_estimator_
```

```python
plt.figure(figsize=(100, 60))
plot_tree(best_desicion_tree, filled=True, feature_names=X_train.columns, class_names=['Not Survived', 'Survived'], rounded=True)
plt.show()
```

```python
best_score = grid_search.best_score_
print(f"Best Score of the decision tree is from GridSearch: {best_score}")

cross_val_scores = cross_val_score(best_desicion_tree, X_train, Y_train, cv=5, scoring='accuracy')
mean_cross_val_score = cross_val_scores.mean()

print(f"Mean Cross-Validation Score of the decision tree from cross validation: {mean_cross_val_score}")
```

```python
param_grid_rf = {
    'n_estimators' : [50,100,150, 200],
    'max_depth' : [3,5,8,10],
    'min_samples_leaf' : [1, 2, 4],
    'min_samples_split' : [2, 5, 10]
}

random_forest = RandomForestClassifier()
grid_search_rf = GridSearchCV(estimator=random_forest, param_grid=param_grid_rf, cv=5, scoring= 'accuracy')
grid_search_rf.fit(X_train, Y_train)
best_rf = grid_search_rf.best_estimator_
```

```python
best_score_rf = grid_search_rf.best_score_
print(f"Best Score of the random forest is from GridSearch: {best_score_rf}")

cross_val_scores_rf = cross_val_score(best_rf, X_train, Y_train, cv=5, scoring='accuracy')
mean_cross_val_score_rf = cross_val_scores_rf.mean()

print(f"Mean Cross-Validation Score of the random forest from cross validation: {mean_cross_val_score_rf}")
```

Random forest algorithm sightly performed better than decision tree based on the mean accuracy from the classification accuracy that is (81.93 > 80.92).

Observations:
1. random forest combines the predictions of many decision trees, which make them more reliable and often gives us better results than single decision tree. This combination of tree helps perform better on unseen data.

2. Random forest add randomness by using different samples of data and different sets of features at each split. This helps the model to capture a wide range of patterns and relationships in the data.

3. Random forest takes more computing power than a single decision tree, as it involes building multiple trees.But it covers for its improved accuracy.

4. Random forests helps in avoiding overfitting which can be an issues as it can memorize the data while training. random forest avoids this by averaging the results of many trees.


Conclusions: Random forest can be chosen generally over the decision trees if you do not care much about the computation and care about only the accuracy. however, the choice should also consider other factors like training time, interpretebility and the type of problem while choosing a model.
